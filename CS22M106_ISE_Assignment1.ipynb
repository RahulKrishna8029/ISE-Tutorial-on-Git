```python
#get a list of all branches in the repository
list(repo.get_branches())
```

```python
#extracting information from a issues
issues_reported=repo.get_issues()
issue_report_info={}
k=0
for ir in issues_reported[:200]:
  #issue title and description
  ir_title=ir.title
  ir_body=ir.body
  #issue labels
  ir_labels=[]
  for label in ir.labels:
    ir_labels.append(label.name)
  #issue comments
  issue_comments=ir.get_comments()
  ir_comments=[]
  for comment in issue_comments:
    ir_comments.append(comment.body)
  #aggregate extracted data into a dictonary
  dict_pr={}
  dict_pr['Issue Number']=ir.number
  dict_pr['Issue Title']=ir.title
  dict_pr['Issue Label']=ir.labels
  dict_pr['Issue state']=ir.state
  dict_pr['Issue Body']=ir.body
  dict_pr['Issue Comments']=ir_comments
  k=k+1
  issue_report_info[k]=dict_pr
```

```python
import pandas as pd
df=pd.DataFrame(issue_report_info)
```
